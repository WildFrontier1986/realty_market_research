### Работа с пропусками и дубликатами

```python
# Удаление строк с пропусками в 'floors_total' и 'locality_name'
df = df.dropna(subset=['floors_total', 'locality_name']).reset_index(drop=True)
```

```python
# Список заменяемых ключевых слов неявных дубликатов
wrong_names = [
    'поселок', 
    'городской посёлок', 
    'посёлок Мурино', 
    'деревня Кудрово', 
    'посёлок при железнодорожной станции Вещево',
    'посёлок при железнодорожной станции Приветнинское'
]
```

```python
# Список слов-заменителей
right_names = [
    'посёлок', 
    'посёлок городского типа', 
    'Мурино', 
    'Кудрово', 
    'посёлок станции Вещево', 
    'посёлок станции Приветнинское'
]
```

```python
# Создание функции для замены дубликатов в df['locality_name']
# 1. Объявление функции rename()
def rename(data, column):
    # 2. Вложенный цикл перебора значений в wrong_names
    for number in range(len(wrong_names)):
        # 2.1 Метод замены дубликатов
        data[column] = data[column].str.replace(wrong_names[number], right_names[number])
    # 2.2 Возвращение функции
    return data[column]
```

```python
# Применение функции rename() к колонке 'locality_name'
# для замены дубликатов с обработкой ошибок try-except
try:
    rename(df, 'locality_name')
    display(df['locality_name'].sort_values().unique())
except:
    print('Проверь алгоритм работы функции "rename"!')
```

```python
# Замена пропусков в 'is_apartment' на булевое значение False
df['is_apartment'] = df['is_apartment'].fillna(False)
```

```python
# Отключение предупреждения 'Mean of empty slice'
# warnings.filterwarnings(action='ignore', message='Mean of empty slice')
```

```python
# Замена пропусков высоты потолков в 'ceiling_height' медианным
# значением из группы по суммарному количеству этажей
df['ceiling_height'] = df.groupby('floors_total', sort=False)['ceiling_height'].apply(lambda x: x.fillna(x.median()))
```

    Данный пропуск заполняю медианным значением по всей колонке 'ceiling_height'.</b></span></div>

```python
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
```

```python
studio_open = ['studio', 'open_plan']
```

```python
# Заполнение пропусков в 'living_area' для студий и квартир со
# свободной планировкой с помощью функции area_drop_nan1()
# 1. Объявление функции area_drop_nan1()
def area_drop_nan1(data, column):
    # 2. Вложенный цикл второго периода для
    # перебора значений в списке studio_open
    for value in studio_open:
        # 2.1. Формирование датафрейма по значению из списка
        data_value = data[data[value] == True]
        # 2.2. Получение медианного значения колонки
        # column в сформированном датафрейме
        data_value_median = data_value[column].median()
        # 2.3. Заполнение пропусков при помощи логической индексации
        data.loc[
            (data[column].isna()) &
            (data[value] == True),
            column
        ] = data_value_median
    return data[column]
```

```python
# Создание списка индексов для группировки методом pivot_table()
indexes_new = ['rooms', 'is_apartment', 'studio', 'open_plan']
```

```python
# Создание списка колонок для расчёта значений в параметре values
values_new = ['living_area', 'kitchen_area']
```

```python
# Заполнение пропусков в 'living_area' и 'kitchen_area' для
# квартир и апартаментов с помощью функции area_drop_nan2()
# 1. Объявление функции area_drop_nan2()
def area_drop_nan2(data, index_list, values_list):
    # 2.1 Формирование сводной таблицы с медианными значениями,
    # вывод сводной таблицы на экран
    data_pivot = data.pivot_table(
        index=index_list,
        values=values_list,
        aggfunc='median'
    )
    display(data_pivot)
    # 2.2 Объединение data с data_pivot
    data_join = data.join(
        data_pivot, 
        on=index_list, 
        rsuffix='_median'
    )
    # 2.3 Заполнение пропусков в колонках 'living_area', 'kitchen_area'
    # значениями из списка fillna_columns
    fillna_columns = ['living_area_median', 'kitchen_area_median']
    data_join.loc[
        data_join[values_new[0]].isna(), 
        values_new[0]
    ] = data_join[fillna_columns[0]]
    data_join.loc[
        data_join[values_new[1]].isna(), 
        values_new[1]
    ] = data_join[fillna_columns[1]]
    # 2.4 Декомпозиция data_join удалением колонок
    # 'living_area_median', 'kitchen_area_median'
    data_join.drop(columns=fillna_columns, inplace=True)
    display(data_join)
    return data_join
```

```python
# Применение функции area_drop_nan2() к датафрейму df и
# к спискам indexes_new и values_new для заполнения пропусков
# с перезаписью датафрейма и обработкой ошибок try-except
try:
    df = area_drop_nan2(df, indexes_new, values_new)
except:
    print('Проверь алгоритм работы функции "area_drop_nan2()"!')
```

```python
# Заполнение пропусков в 'days_exposition' средним
# значением по населённому пункту и количеству комнат
# с помощью функции exposition_drop_nan()
# 1. Объявление функции exposition_drop_nan()
def exposition_drop_nan(data, column):
    # 2. Формирование списка уникальных названий городов
    locality_names = data['locality_name'].unique()
    # 3. Вложенный цикл для перебора значений
    # в списке locality_names
    for value1 in locality_names:
        # 3.1. Формирование датафрейма по значению из списка
        data_value1 = data[data['locality_name'] == value1]
        # 3.2. Формирование уникального списка количества комнат
        rooms_list = data_value1['rooms'].unique()
        # 4. Вложенный цикл второго периода для перебора
        # значений в уникальном списке количества комнат
        for value2 in rooms_list:
            # 4.1. Формирование датафреймов с помощью
            # логической индексации по значению value2
            data_value2 = data_value1[data_value1['rooms'] == value2]
            # 4.2. Получение среднего значения колонки
            # column в сформированном датафрейме
            data_value2_mean = data_value2[column].mean()
            # 4.3. Заполнение пропусков при помощи логической индексации
            data.loc[
                (data[column].isna()) &
                (data['locality_name'] == value1) &
                (data['rooms'] == value2),
                column
            ] = data_value2_mean
    return data[column]
```

```python
# Применение функции exposition_drop_nan() к колонке 'days_exposition'
# для заполнения пропусков с обработкой ошибок try-except
try:
    exposition_drop_nan(df, 'days_exposition')
except:
    print('Проверь алгоритм работы функции "exposition_drop_nan()"!')
```
